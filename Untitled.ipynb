```python
import os
```

```python
all_files = os.listdir()
txt_files = list(filter(lambda x: x.split('.')[-1] == 'txt', all_files))
txt_files.sort()
```

```python
captions = []
for path in txt_files:
    with open(path, 'r') as f:
        captions.append(f.read())
```

```python
len(captions)
```

```python
txt_files
```

```python
import geocoder

places = list(map(lambda x: x.split('.')[0] + ',' + ' CT, USA', captions))

geos = list(map(lambda x: geocoder.google(x, key="AIzaSyC1JMSeR7XeqIH5AB-gY_nqmGNYbzE3lyQ").latlng, places))
```

```python
import csv

with open('data.csv', 'w') as f:
    csv_writer = csv.writer(f)
    for i in range(0, len(captions)):
        csv_writer.writerow([
            txt_files[i],
            txt_files[i].replace('.txt', '.jpg'),
            captions[i].strip().replace('\n', ''),
            captions[i].split('.')[0],
            str(geos[i][0]) if type(geos[i]) == list else '',
            str(geos[i][1]) if type(geos[i]) == list else ''
        ])
```
